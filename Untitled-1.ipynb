H0 = Feature mean has no differen in Target variable
H1 = Feature mean has significant difference in Target variable

```python
#Hypothesis testing

def int_hypothesis(df):
    
    p_val=[]
    for i in int_fea:
        
        pota_1 = df[df['TARGET_Outage']=='Outage'][i]
        pota_0 = df[df['TARGET_Outage']=='Non-Outage'][i]
        stat, p_value=ttest_ind(pota_1, pota_0)
        p_val.append(np.round(p_value,3))
        if p_value <0.1:
            print(f"p_value for {i} is {p_value} is less than significant value 0.1, so we have no enough evidance to prove Null Hypothesis. so we reject the Null Hypotesis")
        else:
            print(f"p_value for {i} is {p_value} we accept the null hypothesis")

    stats_test=pd.DataFrame(columns=['columns','p_value'])
    stats_test['columns']=int_fea
    stats_test['p_value']=p_val
    stats_test.sort_values(by=['p_value'], ascending=True, inplace=True)

    fig=plt.figure(figsize=(15,8))
    ax=sns.barplot(data=stats_test, x='columns',y='p_value')
    plt.title("Features and p_value based on t-Test", size=20, weight='bold')
    for i in ax.patches:
        ax.text(x=i.get_x()+0.5, y=i.get_height(), s=i.get_height(), rotation=90)
    ax.axhline(y=0.1, color='red', ls='--')
    plt.xticks(rotation=90, ha='right')
    ax.text(x=6, y=0.12, s="Significance level: 0.1")
```

```python
int_hypothesis(df.dropna(axis=0))
```

Business_Risk_AOTS - Most data are towards Non_Distruptive & Disruptive. Other category doesn't have required data. lets check the Hypothesis testing to prove the same. 

H0 = There is no relation between the variables
H1 = There are relationship between the valriables.

```python
#Chi Square test

def chi_test(crosstab, fea):
    chi, p_value, dof, _=chi2_contingency(crosstab, correction=True)
    if p_value <0.05:
        print(f"p_value for {fea} is {p_value:e} is less than significant value 0.05, There are relationship between the vairables")
    else:
        print(f"p_value for {fea}  is {p_value:e} we accept the null hypothesis, There are no relationship between the variables")
```

```python

crosstab = pd.crosstab(df['Business_Risk_AOTS'], df['TARGET_Outage'])
chi_test(crosstab,'Business_Risk_AOTS')
```

```python
df1=df[df['Business_Risk_AOTS'].isin(['Non-Disruptive','Disruptive'])]
crosstab=pd.crosstab(df1['Business_Risk_AOTS'],df1['TARGET_Outage'])
chi_test(crosstab,'Business_Risk_AOTS')
```

```python
df1=df[~df['Business_Risk_AOTS'].isin(['Non-Disruptive','Disruptive'])]
crosstab=pd.crosstab(df1['Business_Risk_AOTS'],df1['TARGET_Outage'])
chi_test(crosstab,'Business_Risk_AOTS')
```
