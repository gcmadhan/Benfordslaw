H0 = Feature mean has no differen in Target variable

```python
#Hypothesis testing

def int_hypothesis(df):
    
    p_val=[]
    for i in int_fea:
        
        pota_1 = df[df['TARGET_Outage']=='Outage'][i]
        pota_0 = df[df['TARGET_Outage']=='Non-Outage'][i]
        stat, p_value=ttest_ind(pota_1, pota_0)
        p_val.append(np.round(p_value,3))
        if p_value <0.1:
            print(f"p_value for {i} is {p_value} is less than significant value 0.1, so we have no enough evidance to prove Null Hypothesis. so we reject the Null Hypotesis")
        else:
            print(f"p_value for {i} is {p_value} we accept the null hypothesis")

    stats_test=pd.DataFrame(columns=['columns','p_value'])
    stats_test['columns']=int_fea
    stats_test['p_value']=p_val
    stats_test.sort_values(by=['p_value'], ascending=True, inplace=True)

    fig=plt.figure(figsize=(15,8))
    ax=sns.barplot(data=stats_test, x='columns',y='p_value')
    plt.title("Features and p_value based on t-Test", size=20, weight='bold')
    for i in ax.patches:
        ax.text(x=i.get_x()+0.5, y=i.get_height(), s=i.get_height(), rotation=90)
    ax.axhline(y=0.1, color='red', ls='--')
    plt.xticks(rotation=90, ha='right')
    ax.text(x=6, y=0.12, s="Significance level: 0.1")
```

```python
int_hypothesis(df.dropna(axis=0))
```

H0 = There is no relation between the variables
H1 = There are relationship between the valriables.

```python
#Chi Square test

def chi_test(crosstab, fea):
    chi, p_value, dof, _=chi2_contingency(crosstab, correction=True)
    if p_value <0.05:
        print(f"p_value for {fea} is {p_value:e} is less than significant value 0.05, There are relationship between the vairables")
    else:
        print(f"p_value for {fea}  is {p_value:e} we accept the null hypothesis, There are no relationship between the variables")
    return p_value
```

```python

crosstab = pd.crosstab(df['Business_Risk_AOTS'], df['TARGET_Outage'])
p_val1 = chi_test(crosstab,'Business_Risk_AOTS')
```

```python
df1=df[df['Business_Risk_AOTS'].isin(['Non-Disruptive','Disruptive', 'Non-Disruptive Due to Redundancy'])]
crosstab=pd.crosstab(df1['Business_Risk_AOTS'],df1['TARGET_Outage'])
p_val2= chi_test(crosstab,'Business_Risk_AOTS')
```

```python
if p_val1>p_val2:
    print("Business Risk AOTS feature with only Non-Disruptive, Disruptive & Non-Disruptive Due to Redundancy has more relationship with Targer variable")
else:
    print("Business Risk AOTS feature with only Non-Disruptive, Disruptive & Non-Disruptive Due to Redundancy  has less relationship with Targer variable")
```

```python
#"Business_Risk_AOTS" feature excluding 'Non-Disruptive','Disruptive'
df1=df[~df['Business_Risk_AOTS'].isin(['Non-Disruptive','Disruptive','Non-Disruptive Due to Redundancy'])]
crosstab=pd.crosstab(df1['Business_Risk_AOTS'],df1['TARGET_Outage'])
p_val3=chi_test(crosstab,'Business_Risk_AOTS')
```

```python
if p_val3>p_val2:
    print("Business Risk AOTS feature with only Non-Disruptive, Disruptive & Non-Disruptive Due to Redundancy has more relationship with Targer variable")
else:
    print("Business Risk AOTS feature with only Non-Disruptive, Disruptive & Non-Disruptive Due to Redundancy has less relationship with Targer variable")
```

Data Modeling

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
df.dropna(axis=0, inplace=True)
cat_fea=df.select_dtypes(include=['object','category']).columns.to_list()
int_fea=df.select_dtypes(exclude=['object','category','datetime']).columns.to_list()
cat_fea.remove('TARGET_Outage')
num_pipe = Pipeline([('num_trans',StandardScaler())])
cat_pipe = Pipeline([('cat_trans',OneHotEncoder(drop='first',handle_unknown='ignore', sparse=False))])

preprocess = ColumnTransformer(transformers=[('num',num_pipe,int_fea),('cat',cat_pipe,cat_fea)], remainder='passthrough')
```

```python
X=df.drop(['TARGET_Outage'], axis=1)
y=df['TARGET_Outage'].apply(lambda x: 1 if x=='Outage' else 0)
X_train, X_test, y_train, y_test=train_test_split(X,y, train_size=.80, shuffle=True)
#y_train=y_train.values.flatten()
#y_test=y_test.values.flatten()
```

```python
from sklearn.linear_model import LogisticRegression
import shap
model=LogisticRegression()
ml_pipe = Pipeline([('preprocess',preprocess),('model',model)])
ml_pipe.fit(X_train, y_train)
pred = ml_pipe.predict(X_test)
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
sns.heatmap(confusion_matrix(y_test, pred), annot=True)
print("Confusion Matrix: \n")
print(classification_report(y_test, pred))
print(f"Accuracy score: {accuracy_score(y_test, pred)*100:0.2f}%")
```

```python

shap.initjs()
X_shap = preprocess.fit_transform(X_train)
test_shap = preprocess.transform(X_test)
features = preprocess.get_feature_names_out(input_features=int_fea+cat_fea)
explainer = shap.LinearExplainer(model,preprocess.fit_transform(X_train))
shap_values = explainer.shap_values(test_shap)
shap.summary_plot(shap_values, preprocess.transform(X_test), feature_names=features)

```

```python
from sklearn.ensemble import AdaBoostClassifier
model=AdaBoostClassifier()
ml_pipe = Pipeline([('preprocess',preprocess),('model',model)])
ml_pipe.fit(X_train, y_train)
pred = ml_pipe.predict(X_test)
sns.heatmap(confusion_matrix(y_test, pred), annot=True)
print("Confusion Matrix: \n")
print(classification_report(y_test, pred))
print(f"Accuracy score: {accuracy_score(y_test, pred)*100:0.2f}%")
```
